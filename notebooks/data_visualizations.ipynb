```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib import rcParams
rcParams['font.family'] = 'Lucida Grande'
```

```python
Houston = np.array([[700, 500], 
                    [950, 275], 
                    [1100, 215]])

Galaxy = np.array([[930, 350], 
                 [1050, 150]])

Port = np.array([[800, 220],
                [1000, 375]])

Sea = np.array([[1050, 275]])

Orl = np.array([[850, 120]])


shots = np.array([[700, 500], 
                  [930, 350], 
                  [1050, 150],
                 [1000, 375]])

crosses = np.array([[950, 275], 
                    [1100, 215],
                    [800, 220],
                    [1050, 275],
                    [850, 120]])
```

```python
fig, ax = plt.subplots(figsize=(19, 18))
img = plt.imread('../soccer_field2.png')
ax.imshow(img)
plt.scatter(Houston[:, 0], Houston[:, 1], c='darkorange', s=550, marker='*', edgecolors='orangered', label='HOU Dynamo')
plt.scatter(Galaxy[:, 0], Galaxy[:, 1], c='midnightblue', s=550, marker='*', edgecolors='navy', label='LA Galaxy')
plt.scatter(Port[:, 0], Port[:, 1], c='yellow', s=600, marker='*', edgecolors='forestgreen', label='PHI Union')
plt.scatter(Sea[:, 0], Sea[:, 1], c='forestgreen', s=550, marker='*', edgecolors='darkgreen', label='SEA Sounders')
plt.scatter(Orl[:, 0], Orl[:, 1], c='blueviolet', s=550, marker='*', edgecolors='rebeccapurple', label='ORL City')
ax.set_xticks([])
ax.set_yticks([])
plt.legend(loc='lower left', fontsize='x-large')
plt.title('LAFC Location of Conceded Goals', size=25, pad = 15)
plt.savefig('../lafc_conceded_goals.png', dpi=100);
```
